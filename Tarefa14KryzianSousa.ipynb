**1)** Para este item você deverá usar o arquivo [com dados de força e torque medidos no chão enquanto uma pessoa tenta ficar em pé durante um minuto](dados/balance.csv).

- Faça o gráfico do torque na direção y (My [Nm]) em função do tempo (Time [s]).

- Mostre na tela a média e o desvio-padrão do torque na direção y.

```python
import numpy as np #Comando para importar a biblioteca numpy
import pandas as pd #Comando para importar a biblioteca pandas
import matplotlib.pyplot as plt #Comando para importar a biblioteca matplotlib
dados = pd.read_csv('https://raw.githubusercontent.com/BMClab/BasesComputacionais/master/dados/balance.csv') #Comando para leitura do arquivo csv
```

```python
media = np.mean(dados['My[Nm]']) #Comando para média
print("A média do torque na direção y é " + str(round(media,2))) #Comando para print da média
desvio = dados['My[Nm]'].std() #Comando para desvio padrão
print("O desvio-padrão do torque na direção y é " + str(round(desvio,2))) #Comando para print do desvio padrão
```

```python
plt.figure(figsize=(20,5)) #Comando para fazer o gráfico e definir tamanho
plt.plot(dados['Time[s]'], dados['My[Nm]'], color='red') #Como para plotar dados de acordo com enunciado do exercício
plt.xlabel('Tempo(s)') #Nome da abscissa
plt.ylabel('Torque em y [Nm]') #Nome da ordenada
plt.title('Torque na direcao y em funcao do tempo') #Título do gráfico
plt.xkcd() #Modo cartoon
plt.show() #Comando para mostrar gráfico
```

```python
import numpy as np #Comando para importar biblioteca numpy
import pandas as pd #Comando para importar biblioteca pandas
import matplotlib.pyplot as plt #Comando para importar biblioteca matplotlib
inflacao = pd.read_csv('https://raw.githubusercontent.com/Kryzian/BasesComputacionais/master/aula7/dados/inflacaoMensal.csv') #Comando para leitura do arquivo csv
```

```python
mediana=inflacao.query("Mês == 3")['Inflação'].median() #Comando para cálculo da mediana da inflação
print("A mediana da inflacao mensal do mês de março é " + str(mediana)) #Print do cálculo da mediana

media=inflacao.query("Mês == 3")['Inflação'].mean() #Comando para cálculo da média da inflação
print("\nA Média da inflação do mês de março é " + str(round(media,3))) #Print do cálculo da média da inflação

desvio = inflacao.query("Mês == 3")['Inflação'].std() #Comando para cálculo do desvio da inflação
print("\nO desvio padrão da inflação mensal no mês de março é " + #Print do cálculo do desvio da inflação
      str(round(desvio,3)))
```
