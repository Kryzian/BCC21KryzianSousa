**1)**  Três séries, $v, y$ e $t$ seguem as seguintes regras:
        
$$v[i] = v[i-1] + 0,01.(-9,81)$$
$$y[i] = y[i-1] + 0,01v[i-1]$$
$$t[i] = t[i-1] + 0,01$$
        
Começando com v[0] = 10 m/s, y[0] = 0 m e t[0] = 0 s, calcule os 250 primeiros valores das três séries. 

Faça os gráficos com os valores encontrados de $y$ em função de $t$ e dos valores encontrados de $v$ em função de $t$.

Observação: o cálculo destas três séries é uma simulação computacional para calcular a posição e velocidade de uma partícula lançada na vertical com velocidade inicial de 10 m/s. O método utilizado para realizar esta simulação é conhecido como método de Euler.


```python
import numpy as np #Comando para importar a biblioteca numpy
import matplotlib.pyplot as plt #Comando para importar a biblioteca matplotlib

v = np.zeros((250)) #Comandos para os 250 primeiros valores das três séries
y = np.zeros((250))
t = np.zeros((250))

v[0]=10 #Comandos para variáveis de acordo com enunciado do exerecício
y[0]=0
t[0]=0

for i in range(1, 250): #Comando para repetir o código a cada elemento da lista
  v[i]=v[i-1]+0.01*(-9.81) #Comando para fórmula dada v
  y[i]=y[i-1]+0.01*v[i-1] #Comando para fórmula dada y
  t[i]=t[i-1]+0.1 #Comando para fórmula dada t

plt.figure(figsize=(10,6)) #Comando para fazer o gráfico e definir tamanho
plt.plot(t,y, color='pink') #Comando para definir cor
plt.xlabel('Tempo (s)') #Nome da abscissa
plt.ylabel('Posicao (m)') #Nome da ordenada
plt.title('Posicao em funcao do tempo') #Título do gráfico
plt.grid()  #Comando para linhas de grade
plt.xkcd() #Comando para modo cartoon
plt.show #Comando para mostrar o gráfico
print("\n") #Comando para mostrar resultado

plt.figure(figsize=(10,6)) #Comando para fazer o gráfico e definir tamanho
plt.plot(t,v, color='r') #Comando para definir cor
plt.xlabel('Tempo (s)') #Nome da abscissa
plt.ylabel('Velocidade (m/s)') #Nome da ordenada
plt.title('Velocidade em funcao do tempo') #Título do gráfico
plt.grid() #Comando para linhas de grade
plt.xkcd() #Comando para modo cartoon
plt.show #Comando para mostrar o gráfico
```

**2)** O valor de $\ln(1+x)$ pode ser aproximado por (apenas para $|x|<1$):

$\ln(1+x) = x - \frac{x^2}{2} + \frac{x^3}{3} ... = \displaystyle\sum_{i=1}^n (-1)^{i+1}\frac{x^i}{i} $

$\ln$ é o logaritmo na base $e$ (conhecido como logaritmo natural).

Use a expressão acima com n = 100 para encontrar $\ln(0.5)$.

Compare o resultado com a função do Numpy `np.log(0.5)`.

```python
import numpy as np #Comando para importar biblioteca numpy
n=100 #In(1+x) => In(0.5):=> x=-0.5 
x = -0.5
In = 0

for i in range(1, n+1): #Comando para repetir o código a cada elemento da lista
  if i % 2 == 0: #Comando para executar de acordo com a condição
    In -= x**i/i
  else: #Comando para executar caso contrário
    In += x**i/i

print("Logaritmo utilizando a expressao dada:" + str(In)) #Comando print para str(In)
print("Logaritmo pela funcao np.log:" + str(np.log(0.5))) #Comando print para str(np.log(0.5))
```

**3)** Calcular a aproximação de uma onda quadrada seguindo a seguinte expressão (a expressão abaixo é calculada utilizando uma teoria matemática conhecida como série de Fourier):

$$x(t)  \approx \frac{4}{\pi}\displaystyle \sum_{i = 1}^N \frac{\sin\left[(2i - 1)t\right]}{2i - 1}$$

Faça a aproximação para $0 \leqslant t \leqslant 10$ (utilize um número suficiente de pontos no vetor $t$ para poder observar o gráfico de $x(t)$, algo em torno de 100000 pontos).

Faça a aproximação para a) N = 10, b) N = 100, c) N = 1000 e d) N = 10000. Coloque todas as aproximações em um único gráfico, com a aproximação de $x(t)$ em função de $t$.

Dica: se você estiver com dificuldade, faça primeiro o gráfico considerando i = 1, sem nenhuma somatória (N = 1). Depois disso, coloque o código que você tiver escrito dentro de um for ou while para variar o valor de i e fazer a somatória.

```python
import numpy as np #Comando para importar a biblioteca numpy
import matplotlib.pyplot as plt #Comando para importar a biblioteca matplotlib
t = np.linspace(0, 10.0001, 100000) #Comando para sequencia de números espaçados
x = np.zeros((len(t))) #Comando para criacao de arranjo 
```

```python
def Fourier (x, N):
  for i in range(1, len(t)):#Comando para colocar condições
    soma = 0
    for h in range(1, N): #Comando para colocar condições
      soma += np.sin((2*h - 1)*t[i])/(2*h - 1) #De acordo com a fórmula dada
    x[i] = 4/np.pi * soma
  return x #Comando para mostrar o resultado
```

```python
plt.figure(figsize=(15,5)) #Comando para fazer o gráfico
plt.plot(t, Fourier(x,10)) #Comandos para plotar de acordo com Fourier X,10
plt.plot(t, Fourier(x,100)) #Comandos para plotar de acordo com Fourier X,100
plt.plot(t, Fourier(x,1000)) #Comandos para plotar de acordo com Fourier X,1000
plt.plot(t, Fourier(x,10000)) #Comandos para plotar de acordo com Fourier X,10000
plt.xlabel('t') #Nome da abscissa
plt.ylabel('x(t)') #Nome da ordenada
plt.title('Serie de Fourier para calculo de uma onda quadrada') #Comando para título
plt.xkcd #Comando para modo cartoon
plt.show() #Comando para mostrar gráfico
```
