**1)** O índice de massa corporal (IMC) de uma pessoa é calculado por:
    
$$IMC = \frac{m}{h^2} $$
    
em que $m$ é a massa corporal, em kg, e $h$ é a altura, em metros.

- 25 a 30 kg/m$^2$ -- Acima do peso	
- 30 a 35 kg/m$^2$ -- Obesidade Grau I	
- 35 a 40 kg/m$^2$ -- Obesidade Grau II	
- maior que 40 kg/m$^2$ -- Obesidade Grau III	

Escreva um código em Python para calcular o IMC e mostrar a classificação do IMC. Teste para:

**2)** A acidez $A(x)$ de uma solução de hidróxido de magnésio em ácido clorídrico, sob certas condições experimentais, é dada pela equação:

$$A(x) = x^3 + 3x^2 − 54$$
na qual $x$ é a concentração de íons hidrônio. 

Pede-se:

1) Gere o gráfico de $A(x)$ em função de $x$ para 0 $\leqslant x \leqslant 8$.


```python
import numpy as np #Numpy para criar vetores
x = np.linspace(0,8,100) #Comando para criar vetores especificando os limites e o número de elementos
print ("Estes são os valores para 0≤ x ≤ 8:", x,) #Comando para mostrar variáveis de x
```

```python
a=x**3+(3*x**2)-54 #Aplicando variáveis X na fórmula
print(a) #Comando para mostrar resultado
```

```python
import matplotlib.pyplot as plt
plt.figure() #Comando para fazer o gráfico
plt.plot(x, a) #Os vetores contendo os dados da abscissa e os dados da ordenada
plt.title("Solucao de MgOH2 em HCl") #Título do gráfico
plt.xlabel('Concentracao de hidronio') #Nome da abscissa
plt.ylabel('Acidez') #Nome da ordenada
plt.xkcd() #Comando para deixar o gráfico no modo cartoon
plt.show() #Comando para mostrar o gráfico
```

**3)**  Faça o gráfico de:
    $$f(x)=|x−2|+|2x+1|−x−6$$
    
 Faça o gráfico com $x$ variando de -10 a 10.
 
 Mostre o menor valor de $x$ tal que $f(x)>0$ e $x>0$.

```python
X = np.arange(-10,10,1)
print(X)
```

```python
f=abs(X-2)+abs(2*X+1)-X-6
print(f)
```

```python
c = 0 #variável contadora para pegar só o primeiro valor em comum entre os dois vetores
for x in np.where(X>0)[0]: #verificando só os valores X>0 e f>0
  for y in np.where(f>0)[0]: #verificando só os valores X>0 e f>0
    if x == y and c==0: #se x for igual y (os dois maiores que zero) E for o primeiro valor me retorna ele
      x_ponto = X[x]
      y_ponto = f[y]
      c=c+1
```

```python
import matplotlib.pyplot as plt
plt.figure() #Comando para fazer o gráfico
plt.plot(X, f) #Os vetores contendo os dados da abscissa e os dados da ordenada
plt.title("Grafico da funcao") #Título do gráfico
plt.xkcd() #Comando para deixar o gráfico no modo cartoon
plt.plot(x_ponto, y_ponto, "o") #fazer o plot e marcar o ponto
ax = plt.gca() #pega as informações da moldura/linhas ao redor do gráfico
ax.spines['bottom'].set_position(('data',0)) #muda a posição da linha de baixo (eixo x) pra 0 (para facilitar a visualização)
ax.spines['left'].set_position(('data',0)) #muda a posição da linha da esquerda (eixo y) pra 0 (para facilitar a visualização)
plt.xlabel('X') #Nome da abscissa
plt.ylabel('f(x)') #Nome da ordenada
plt.show() #Comando para mostrar o gráfico
```

**4)** Faça o gráfico da seguinte função:

$$f(x) = x^2 – \mathrm{sen}(0.784x^2) – 2$$

Mostre as raízes da função (i.e. os valores de $x$ para os quais $f(x) = 0$).

```python
import numpy as np
x = np.arange(-10, 10, 0.001) #Menor intervalor possível p/ tentar encontrar a raíz
print("Valores para x",x,) #Comando para mostrar valores de X
f=x**2-np.sin(0.784*x**2)-2 #Fórmula dada
print("Valores para f",f,) #Comando para mostrar valores de Y
```

O F(X) mais próximo de 0 é o 0,12192304 que bate o o X=1,700000 (raíz)

```python
import matplotlib.pyplot as plt
plt.figure() #Comando para fazer o gráfico
plt.plot(x, f) #Os vetores contendo os dados da abscissa e os dados da ordenada
f=x**2-np.sin(0.784*x**2)-2 #Fórmula dada
raizes = x[np.isclose(f,0, atol=0.002)]
print("Essas são as raízes",raizes,)
plt.title("Grafico da funcao") #Título do gráfico
plt.xlabel('x') #Nome da abscissa
plt.ylabel('F(x)') #Nome da ordenada
ax.spines['bottom'].set_position(('data',0)) #muda a posição da linha de baixo (eixo x) pra 0 (para facilitar a visualização)
ax.spines['left'].set_position(('data',0)) #muda a posição da linha da esquerda (eixo y) pra 0 (para facilitar a visualização)
plt.xkcd() #Comando para deixar o gráfico no modo cartoon
plt.show() #Comando para mostrar o gráfico
```
