- Escolha um time e faça o histograma da distribuição do público nos jogos em que este time foi o mandante.

- Calcule qual foi o público médio nos jogos em que o time escolhido foi o mandante.

```python
import pandas as pd #Comando para importar pandas
import matplotlib.pyplot as plt #Comando para importar matplotlib
tabela = pd.read_csv('https://raw.githubusercontent.com/Kryzian/BasesComputacionais/master/aula7/dados/tabelaBrasileirao2018.csv') #Comando para leitura do arquivo csv

plt.figure() #Comando para fazer o gráfico
plt.hist([tabela.query("Mandante == 'Bahia'")['Público']]) #Comando para plotar histograma
plt.xlabel('Publico') #Nome da abscissa
plt.xkcd() #Comando para deixar o gráfico no modo cartoon
plt.show() #Comando para mostrar o gráfico

print("Média do publico quando Bahia é mandante" ,[tabela.query("Mandante == 'Bahia'")['Público'].median()]) #Comando print
```
