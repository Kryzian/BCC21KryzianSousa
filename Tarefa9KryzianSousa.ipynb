```python
import pandas as pd #Comando para importar a biblioteca pandas
import numpy as np #Comando para importar a biblioteca numpy
df = pd.read_csv('https://raw.githubusercontent.com/BMClab/BasesComputacionais/master/dados/analfabetismo.csv') #Comando para leitura do arquivo csv
df
```

```python
import matplotlib.pyplot as plt #Comando para importar matplotlib
df.plot(kind = 'scatter', x = 'Ano', y = 'São Paulo') #Comando para plotar gráfico de acordo com enunciado do exercício
plt.title("Analfabetismo em SP") #Título do gráfico
plt.xkcd() #Comando para deixar o gráfico no modo cartoon
plt.show()
```
