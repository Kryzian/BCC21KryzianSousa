- **1)** Uma bola é lançada ao ar. Suponha que sua altura $h(t)$, em metros, t segundos após o lançamento, seja:
$$h(t) = −t^2 + 4t$$
    
a) Faça o gráfico da altura em função do tempo. Não se esqueça de indicar o significado de cada eixo, colocando a unidade da altura e do tempo.  Sugestão: fazer o gráfico com o tempo entre 0 s e 5 s com incrementos de 0.1 s.



a) Gráfico da altura em função do tempo.

```python
import numpy as np
t = np.array([0,1, 2, 3, 4, 5])
```

```python
t = np.arange(0,5,0.1) #Valores indicando os limites de acordo com o intervalo de 0.1 entre os elementos
print("Estes são os limites de acordo com o intervalo 0.1 entre os elementos", t,)
```

```python
h=-t**2+4*t #Cálculo da altura de acordo com a fórmula dada
print("Estas são as alturas de acordo com o tempo:", h,)
```

```python
import matplotlib.pyplot as plt
plt.figure() #Comando para fazer o gráfico
plt.plot(t, h) #Os vetores contendo os dados da abscissa e os dados da ordenada
plt.title("Grafico da altura em funcao do tempo") #Título do gráfico
plt.xlabel('Tempo') #Nome da abscissa
plt.ylabel('Altura') #Nome da ordenada
plt.xkcd()
plt.show() #Comando para mostrar o gráfico
```


b) Calcule a altura máxima atingida pela bola
    
    

```python
m = np.max(h)
print(f'A altura máxima atingida pela bola foi {m}', 'm')
```
